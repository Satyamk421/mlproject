```python
df['total score'] = df['math_score']+df['reading_score']+df['writing_score']
df['average'] = df['total score']/3
```

```python
reading_full = df[df['reading_score']==100]['average'].count()
math_full = df[df['math_score']==100]['average'].count()
writing_full = df[df['writing_score']==100]['average'].count()
```

```python
fig,axs=plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='average',bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='average',bins=30,kde=True,hue=df['gender'])
```

```python
fig,axs=plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='total score',bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='total score',bins=30,kde=True,hue=df['gender'])
```

```python


fig, axs = plt.subplots(1, 3, figsize=(25, 6))

# Plot 1: Overall average distribution by lunch type
sns.histplot(data=df, x='average', kde=True, hue='lunch', ax=axs[0])
axs[0].set_title('Average Scores by Lunch (All Students)')

# Plot 2: Female students
sns.histplot(data=df[df['gender']=='female'], x='average', kde=True, hue='lunch', ax=axs[1])
axs[1].set_title('Average Scores by Lunch (Female)')

# Plot 3: Male students
sns.histplot(data=df[df['gender']=='male'], x='average', kde=True, hue='lunch', ax=axs[2])
axs[2].set_title('Average Scores by Lunch (Male)')

plt.tight_layout()
plt.show()
```
